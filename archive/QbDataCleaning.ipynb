```python
import pandas as pd
import numpy as np
import nfl_data_py as nfl
```

```python
years = list(range(2006, 2022, 1))
```

```python
qbr_data = []
for year in years:
    data = nfl.import_qbr([year])
    data = data[data.season_type == 'Regular']
    #data = data.drop(['season_type', 'game_week', 'team_abb', 'name_short', 'exp_sack', 'penalty', 'sack'], axis=1)
    data = data[['season', 'name_display', 'team', 'rank', 'qbr_total', 'pts_added', 'qb_plays', 'epa_total', 'pass', 'run', 'qbr_raw', 'qualified', 'player_id', 'name_first', 'name_last', 'headshot_href']]
    qbr_data.append(data)
```

```python
qbr_data[0]
```

```python
qb_season_data = []
for year in years:
    data = nfl.import_seasonal_data([year])

    data = data[['player_name', 'season', 'completions', 'attempts', 'passing_yards', 'passing_tds',
                 'interceptions', 'passing_air_yards', 'passing_yards_after_catch', 'passing_first_downs',
                 'passing_epa', 'dakota', 'carries', 'rushing_yards', 'rushing_tds', 'rushing_epa',
                 'fantasy_points_ppr', 'pacr', 'games', 'ppr_sh', 'player_id']]
    qb_season_data.append(data)
```

```python
qb_season_data[-1]
```

```python
id_data = nfl.import_ids(ids=['gsis', 'espn'])
id_data = id_data[['name', 'position', 'birthdate', 'height', 'weight', 'college', 'draft_year', 'draft_round', 'draft_pick', 'gsis_id', 'espn_id']]
id_data = id_data[id_data.position=='QB']
```

```python
id_data
```

```python
for i in range(len(qbr_data)):
    qbr_data[i] = qbr_data[i].merge(id_data, how='inner', left_on='player_id', right_on='espn_id', suffixes=('_1', '_2'))
```

```python
qbr_data[15]
```

```python
qb_data = []
for i in range(len(qbr_data)):
    qb_data.append(qbr_data[i].merge(qb_season_data[i], how='left', left_on='gsis_id', right_on='player_id', suffixes=('_1', '_2')))
```

```python
qb_data[15]
```

```python
for i in reversed(range(2, len(qb_data))):
    qb_data[i] = qb_data[i].merge(qb_data[i-1],
                                          left_on='espn_id',
                                          right_on='espn_id',
                                          suffixes=('', '_n-1'))
    qb_data[i] = qb_data[i].merge(qb_data[i-2],
                                          left_on='espn_id',
                                          right_on='espn_id',
                                          suffixes=('', '_n-2'))
for i in range(len(qb_data) - 1):
    qb_data[i] = qb_data[i].merge(qb_data[i + 1][['fantasy_points_ppr', 'ppr_sh', 'espn_id']],
                                  left_on='espn_id',
                                  right_on='espn_id',
                                  suffixes=('', '_next_year'))
```

```python
data = nfl.import_weekly_data([2021])
data = data[data.season_type == 'REG']
names = data[['player_id', 'player_name', 'season']]
names = names.drop_duplicates(subset='player_id')
data = data.drop(columns=['week', 'season', 'rushing_2pt_conversions', 'receptions', 'targets', 'receiving_yards', 'receiving_tds', 'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
                          'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa', 'receiving_2pt_conversions', 'special_teams_tds', 'fantasy_points_ppr', 'racr', 'target_share', 'air_yards_share', 'wopr'])
games = data.groupby('player_id').count()
games['games'] = games['player_name']
games = games['games']
grouped = data.groupby('player_id').sum()

qb_data = names.join(grouped, on='player_id', how='inner')
qb_data = qb_data.join(games, on='player_id', how='inner')
```
